```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
```

```python
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
import seaborn as sns
sns.heatmap(dataset.isnull())
```

```python
import math
meanTestScore1=math.floor(dataset['Age'].median())
meanTestScore2=math.floor(dataset['Tenure'].median())
meanTestScore3=math.floor(dataset['Balance'].median())
meanTestScore4=math.floor(dataset['NumOfProducts'].median())
meanTestScore5=math.floor(dataset['HasCrCard'].median())
meanTestScore6=math.floor(dataset['IsActiveMember'].median())
meanTestScore7=math.floor(dataset['EstimatedSalary'].median())
meanTestScore8=math.floor(dataset['Exited'].median())
meanTestScore9=math.floor(dataset['CreditScore'].median())

dataset['Age']=dataset['Age'].fillna(meanTestScore1)
dataset['Tenure']=dataset['Tenure'].fillna(meanTestScore2)
dataset['Balance']=dataset['Balance'].fillna(meanTestScore3)
dataset['NumOfProducts']=dataset['NumOfProducts'].fillna(meanTestScore4)
dataset['HasCrCard']=dataset['HasCrCard'].fillna(meanTestScore5)
dataset['IsActiveMember']=dataset['IsActiveMember'].fillna(meanTestScore6)
dataset['EstimatedSalary']=dataset['EstimatedSalary'].fillna(meanTestScore7)
dataset['Exited']=dataset['Exited'].fillna(meanTestScore8)
dataset['CreditScore']=dataset['CreditScore'].fillna(meanTestScore9)

dataset['Gender']=dataset['Gender'].fillna('male')
```

```python
dataset.head(10)
```

```python
dataset=dataset.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 'columns')
```
