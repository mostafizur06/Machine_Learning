```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('student_dataset.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
columns = ['Study(hr)<=10', 'Mental Stess(<6)', 'Social Media active (hr<10)', 'Group study(hr<10)', 'CGPA']

for column in columns:
    median = math.floor(df[column].median())
    df[column] = df[column].fillna(median)

df.head()
```

```python
df.isnull().sum()
```

# Predict the CGPA when CGPA is dependent only on Study Hour

```python
y_CGPA = df['CGPA']
x_StudyHr = df['Study(hr)<=10']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_StudyHr, y_CGPA, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train = np.array(X_train).reshape(-1, 1)
X_test = np.array(X_test).reshape(-1, 1)
```

```python
reg = LinearRegression().fit(X_train, Y_train)
reg.score(X_test,Y_test)
```

## Question 3

```python
df['CGPA'] = np.where(df['CGPA'] >= 3.75, 'PASS', 'Fail')
```

```python
df.head()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['CGPA']= label_encoder.fit_transform(df['CGPA'])
df.head()
```

```python
x3=df.drop(['Mental Stess(<6)', 'Group study(hr<10)', 'CGPA'], axis=1)
y3 = df['CGPA']
```

```python
y3
```

```python
from sklearn.model_selection import train_test_split
X_train3, X_test3, y_train3, y_test3 = train_test_split(x3, y3, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(X_train3, y_train3)
clf.score(X_test3, y_test3)
```
