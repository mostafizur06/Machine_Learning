```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
dataset = pd.read_csv('Lab.csv')
```

```python
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
import math
meanAge=math.floor(dataset['Age'].mean())
meanAge
```

```python
dataset['Age']=dataset['Age'].fillna(meanAge)
```

```python
dataset.head(10)
```

```python
meanPurchaseAmount=math.floor(dataset['Purchase\nAmount'].mean())
meanPurchaseAmount
```

```python
dataset['Purchase\nAmount']=dataset['Purchase\nAmount'].fillna(meanPurchaseAmount)
```

```python
dataset.head(10)
```

```python
meanLoyaltyPoint=math.floor(dataset['Loyalty\nPoints'].mean())
meanLoyaltyPoint
```

```python
dataset['Loyalty\nPoints']=dataset['Loyalty\nPoints'].fillna(meanLoyaltyPoint)
```

```python
dataset.head(10)
```

```python
colums=['Gender','Product\nCategory','Payment\nMethod','Membership\nLevel']
colums
```

```python
for co1 in colums:
  mode_value=dataset[co1].mode()[0]
  dataset[co1].fillna(mode_value, inplace=True)
```
